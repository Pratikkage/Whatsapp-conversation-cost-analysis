```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Calculate Average Cost per Conversation: Calculate the average cost per conversation across different markets and services.
# Identify Highest and Lowest Costs: Find the markets and services with the highest and lowest costs per conversation.
# Visualize Cost Trends: Visualize the cost trends for different services across markets.
# Identify Market Trends: Identify any trends or patterns in costs across different markets.
# Compare Costs Across Services: Compare the costs per conversation for marketing, utility, authentication, and other services.
```

```python
df = pd.read_csv('398966882_3517923138456740_5850766363335151232_n.csv')
```

```python
df.head()
```

```python
# Calculate Average Cost per Conversation: Calculate the average cost per conversation across different markets and services.
```

```python
df['Avg_cost'] = df[['Marketing','Utility','Authentication']].mean(axis=1)
df.head()
```

```python
# Identify Highest and Lowest Costs: Find the markets and services with the highest and lowest costs per conversation.
```

```python
max(df['Avg_cost'])
```

```python
min(df['Avg_cost'])
```

```python
# Visualize Cost Trends: Visualize the cost trends for different services across markets.
```

```python
plt.figure(figsize = (10,6))
plt.bar(df['Market'],df['Avg_cost'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Market')
plt.ylabel('Average cost')
plt.show()
```

```python
# Identify Market Trends: Identify any trends or patterns in costs across different markets.
```

```python
# Compare Costs Across Services: Compare the costs per conversation for marketing, utility, authentication, and other services.
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Market'],df['Avg_cost'], color = 'Red')
plt.bar(df['Market'],df['Utility'])
plt.legend()
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Market'],df['Avg_cost'])
plt.bar(df['Market'],df['Authentication'])
plt.legend()
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Market'],df['Avg_cost'],color = 'Purple')
plt.bar(df['Market'],df['Service'])
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.show()
```
